```python
import requests
import pandas as pd
import io
import pystow
import json
```

```python
url = "https://scicrunch.org/php/data-federation-csv.php"
params = {
    "orMultiFacets": "true",
    "sortField": "Proper Citation",
    "sortAsc": "true",
    "q": "%2A",
    "count": "50",
    "nifid": "nlx_144509-1",
    "exportType": "data",
}
```

```python
api_key = pystow.get_config("scicrunch", "api_token")

index = "RIN_Tool_pr"
url = f"https://scicrunch.org/api/1/elastic/{index}/_search"
query = {
    "size": 20,
    "from": 0,
    "query": {
        "bool": {
            "must": [
                {
                    "query_string": {
                        "fields": ["*"],
                        "query": "antibody",
                        "type": "cross_fields",
                        "default_operator": "and",
                        "lenient": "true",
                    }
                }
            ],
            "should": [
                {"match": {"[item.name](http://item.name/)": {"query": "antibody", "boost": 20}}}
            ],
        }
    },
}

res = requests.get(url, params={"api_key": api_key}, data=json.dumps(query))
res_json = res.json()

# curl -X GET "localhost:9200/_search?pretty" -H 'Content-Type: application/json' -d'


with open("results.json", "w") as file:
    json.dump(res_json, file, indent=2)

res_json
```

```python
h1 = res_json["hits"]["hits"][0]
print(h1["_id"])

h1["_source"]
```
